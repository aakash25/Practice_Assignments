## 1. In Iris DataSet , plot a scatter plot differentiating each classifier with a boundary

```python
irisdf= pd.read_csv("C:\\7\\Batch7\\MatPlotLib\\iris.csv")
irisdf.head()
```

## 2. In wine quality dataset, plot a correlation matrix for all numerical variables and find out which columns are highly correlated to each other.

```python
winequality_df= pd.read_csv("C:\\7\\Batch7\\MatPlotLib\\winequality-red.csv")
winequality_df.describe()
```

```python

correl= winequality_df.corr()
plt.figure(figsize=(12,10))
sns.heatmap(correl,annot=True,cmap=plt.cm.Reds)
plt.show()
```

```python
#Correlation with output variable
cor_target = abs(correl['quality'])
print(cor_target)
#Selecting highly correlated features
relevant_features = cor_target[cor_target>0.45]
relevant_features
```

## 9. In the diabetes dataset, plot a pie-chart showing % of diabetes and % of non-diabetes.

```python

# Data to plot
labels = 'Healthy', 'Diabetic'
sizes = [dib_df['Outcome'][dib_df['Outcome']==0].count(), dib_df['Outcome'][dib_df['Outcome']==1].count()]
colors = ['yellowgreen', 'red']
explode = (0.1, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')
plt.show()
```
